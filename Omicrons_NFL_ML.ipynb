```python
#Pandas Read File - CSV
# Dependencies and Setup
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
# File to Load
file_to_load = "Data/nfl_game_scores.csv"
# Read File and store into Pandas data frame
data = pd.read_csv(file_to_load)
```

```python
#Get last years
last_10y = data[(data['schedule_season'] > 2009)]
```

```python
last_10y.head()
```

```python
#Get Stats
last_10y.describe()
```

```python
#Create a column with a flag for Local or Visit - Patriots
last_10y['home_or_away'] = np.where(last_10y['team_home'] == 'New England Patriots', 1, 0)
```

```python
last_10y
```

```python
#Filter for an specific team - Patriots
last_10y_byteam = last_10y.loc[(last_10y["team_home"] == "New England Patriots") | (last_10y["team_away"] == "New England Patriots")]
```

```python
last_10y_byteam.head()
```

```python
#Select specific columns
final = last_10y_byteam.filter(["team_home","team_away" , "score_home","score_away" ,"weather_temperature", "home_or_away"])
```

```python
final.head()
```

```python
#Add new column with score
final['score'] = np.where(final['team_away'] == 'New England Patriots', final['score_away'], final['score_home'])
```

```python
final.head()
```

```python
#Fill blanks in weather with a the mean of column
final = final.fillna(final.mean())
```

```python
#final.to_excel('final2.xlsx')
```

```python
final.info()
```

```python
final['home_or_away'] = final['home_or_away'].astype('float64')
```

```python
final.info()
```

```python
#Select data for the model
X = pd.DataFrame(final, columns = ['weather_temperature', 'home_or_away'])
y = pd.DataFrame(final, columns=['score'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#Build model
lin_reg_mod = LinearRegression()
lin_reg_mod.fit(X_train, y_train)
pred = lin_reg_mod.predict(X_test)
```

```python
#Calculate the R2_Score between the actual and prediction
test_set_r2 = r2_score(y_test, pred)
#Check if it is closer to 1
print(test_set_r2)
```

```python
df_results = y_test
df_results['Predicted'] = pred
df_results['Residuals'] = abs(df_results['score']) - abs(df_results['Predicted'])
print(df_results)
```
